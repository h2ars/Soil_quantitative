## Loading Data


```python
X_data, X_sg, y_data, X_pca = np.load("combined_data.npz").values()
```

```python
X_columns = [f"W{x}" for x in np.arange(350, 2501, 10)]
y_columns = ["PHH2O", "PHKCL", "CA", "MG", "NA", "K", "CLAY", "ORGC"]
```

## Preprocessing / EDA


```python
corr_mat = pd.DataFrame(
    np.hstack([X_sg, y_data]), columns=X_columns + y_columns).corr()

plt.figure(figsize=(18, 12))

plt.subplot(2, 1, 1)
sns.heatmap(corr_mat.iloc[216:, :-8])

plt.subplot(2, 1, 2)
sns.heatmap(corr_mat.iloc[216:, -8:])
plt.show()
```

```python
plt.figure(figsize=(18, 6))

for sp in X_data:
    plt.plot(np.arange(350, 2501, 10), sp)

plt.xlabel("Wavelength (in nm)")
plt.ylabel("Reflectance Spectra")
plt.show()
```

```python
plt.figure(figsize=(18, 6))

for sp in X_sg:
    plt.plot(np.arange(350, 2501, 10), sp)

plt.xlabel("Wavelength (in nm)")
plt.ylabel("SG Values")
plt.show()
```

```python
plt.figure(figsize=(18, 6))

for sp in X_pca:
    plt.plot(sp)

plt.xlabel("Wavelength (in nm)")
plt.ylabel("SG Values")
plt.show()
```
